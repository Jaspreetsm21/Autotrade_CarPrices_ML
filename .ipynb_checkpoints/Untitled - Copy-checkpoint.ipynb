```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

Car_Title =[]
Car_Price = []
Car_Year = []
Car_Reg = []
Car_Type_Door = []
Car_Mileage = []
Car_engine_size = []
Car_Horse_PW = []
Car_Type_Vehicle = []
Car_Filling_in = []

f = ["ABARTH", "AC", "AIXAM", "ARIEL", "ASTON%20MARTIN", "AUDI"] #open("make.txt", "r")
for pg in range(1, 100):  # here I tested it with 9 pages for speed sake
    #for make in f:
    req_url = 'https://www.autotrader.co.uk/car-search?sort=distance&postcode=sl25je&radius=300&onesearchad=New&onesearchad=used&page=' + str(pg) + "'"        
    #'https://www.autotrader.co.uk/car-search?postcode=SL2%205JE&radius=200&homeDeliveryAdverts=include&year-to=2020&onesearchad=Used&onesearchad=Nearly%20New&onesearchad=New&advertising-location=at_cars&page=1'
    req = urlopen(req_url)
    page_html = req.read()
    req.close()
    page_soup = BeautifulSoup(page_html, "html.parser")
    # BeautifulSoup.findAll is deprecated use find_all instead
    listings = page_soup.find_all("li", {"class": "search-page__result"})
    for listing in listings:
        information_container = listing.find("div", {"class": "information-container"})
        title_container = information_container.find("a", {
            "class": "js-click-handler listing-fpa-link tracking-standard-link"})
        title = title_container.text
        price = listing.find("div", {"class": "vehicle-price"}).text
        #Vehcile Information
        Reg_Year = listing.find("ul", {"class": "listing-key-specs"}).text.split()
        #Year
        if Reg_Year[0] == 'CAT':
            Year_Register= Reg_Year[4]
        else:
            Year_Register = Reg_Year[0]
        #Mar vs Sept   
        if Reg_Year[0] == 'CAT':
            Month_Register= Reg_Year[5].strip('(')
        else:
            Month_Register = Reg_Year[1].strip('(')
            
        #Number of Doors   
        if Reg_Year[0] == 'CAT':
            Type_Door= Reg_Year[7]
        else:
            Type_Door = Reg_Year[3]
        #Milleage 
        if Reg_Year[0] == 'CAT':
            milleage = Reg_Year[8]
        else:
            milleage  = Reg_Year[4]
        #engine size 
        if Reg_Year[0] == 'CAT':
            engine_size = Reg_Year[10]
        else:
            engine_size   = Reg_Year[6]
        #horse power 
        if Reg_Year[0] == 'CAT':
            Horse_PW = Reg_Year[11]
        elif len(Reg_Year) == 9:
            Horse_Pw = None
        else:
            Horse_PW   = Reg_Year[7]
        #Manual vs Auto 
        if Reg_Year[0] == 'CAT':
            Type_Vehicle = Reg_Year[12]
        elif len(Reg_Year) == 9:
            Type_Vehicle = Reg_Year[7]                     
        else:
            Type_Vehicle   = Reg_Year[8]
        #Petrol vs Diesel 
        if Reg_Year[0] == 'CAT':
            Filling_in = Reg_Year[13]
        elif  len(Reg_Year) == 9:
            Filling_in   = Reg_Year[8]
        else:
            Filling_in   = Reg_Year[9]

        #print("make:", make)
#         print("title:", title)
#         print("price:", price)
#         print("Reg:",Reg_Year)
#         print("len",len(Reg_Year))
#         print("Year:",Year_Register)
#         print(Month_Register.strip('('))
#         print("Type_Door:",Type_Door)
#         print("milleage:",milleage)
#         print("engine_size:",engine_size)
#         print("Horse_PW:",Horse_PW)
#         print("Type_Vehicle:",Type_Vehicle)
#         print("Filling_in:",Filling_in)
#         print(req_url)
        
        Car_Title.append(title)
        Car_Price.append(price)
        Car_Year.append(Year_Register)
        Car_Reg.append(Month_Register)
        Car_Type_Door.append(Type_Door)
        Car_Mileage.append(milleage)
        Car_engine_size.append(engine_size)
        Car_Horse_PW.append(Horse_PW)
        Car_Type_Vehicle.append(Type_Vehicle)
        Car_Filling_in.append(Filling_in)
```
